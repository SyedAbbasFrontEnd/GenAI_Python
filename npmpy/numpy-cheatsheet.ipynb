# NumPy Visual Cheat Sheet

## 25 Most Useful NumPy Methods

| Category                | Method                | What It Does            | Example                     |
| ----------------------- | --------------------- | ----------------------- | --------------------------- |
| **Array Creation**      | `np.array()`          | Create NumPy array      | `np.array([1,2,3])`         |
|                         | `np.zeros()`          | Array filled with zeros | `np.zeros((3,3))`           |
|                         | `np.ones()`           | Array filled with ones  | `np.ones((2,4))`            |
|                         | `np.arange()`         | Range of numbers        | `np.arange(0,10,2)`         |
|                         | `np.linspace()`       | Evenly spaced values    | `np.linspace(0,1,5)`        |
| **Array Properties**    | `.shape`              | Dimensions of array     | `arr.shape`                 |
|                         | `.ndim`               | Number of dimensions    | `arr.ndim`                  |
|                         | `.size`               | Total elements          | `arr.size`                  |
|                         | `.dtype`              | Data type               | `arr.dtype`                 |
| **Indexing & Slicing**  | `arr[i]`              | Access element          | `arr[0]`                    |
|                         | `arr[:,0]`            | Select column           | `arr[:,0]`                  |
|                         | `arr[1:5]`            | Slice array             | `arr[1:5]`                  |
| **Math Operations**     | `np.sum()`            | Sum of elements         | `np.sum(arr)`               |
|                         | `np.mean()`           | Average value           | `np.mean(arr)`              |
|                         | `np.max()`            | Maximum value           | `np.max(arr)`               |
|                         | `np.min()`            | Minimum value           | `np.min(arr)`               |
|                         | `np.std()`            | Standard deviation      | `np.std(arr)`               |
| **Sorting & Filtering** | `np.sort()`           | Sort array              | `np.sort(arr)`              |
|                         | `np.where()`          | Conditional filtering   | `np.where(arr>10)`          |
| **Reshaping**           | `reshape()`           | Change array shape      | `arr.reshape(3,4)`          |
| **Combining Arrays**    | `np.vstack()`         | Stack vertically        | `np.vstack((a,b))`          |
|                         | `np.hstack()`         | Stack horizontally      | `np.hstack((a,b))`          |
| **Matrix Operations**   | `np.dot()`            | Matrix multiplication   | `np.dot(A,B)`               |
| **Random Numbers**      | `np.random.rand()`    | Random floats           | `np.random.rand(3,3)`       |
|                         | `np.random.randint()` | Random integers         | `np.random.randint(1,10,5)` |

# Mental Map for Students

You can think NumPy as **5 core capability groups**.

```
Create Arrays
     ↓
Inspect Arrays
     ↓
Index & Slice Data
     ↓
Mathematical Operations
     ↓
Reshape & Combine Arrays
```
